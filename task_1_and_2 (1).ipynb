```python
def fix_date_like_columns(df, cols):
    for c in cols:
        if c in df.columns:
            if np.issubdtype(df[c].dtype, np.datetime64):
                try:
                    df[c] = df[c].dt.day.astype("Int64")
                except:
                    df[c] = df[c].astype("Int64")
    return df


check_cols = ["weekday", "workingday", "holiday"]

hour_df = fix_date_like_columns(hour_data, check_cols)
day_df = fix_date_like_columns(day_data, check_cols)
for df, name in [(hour_data, "hour_data"), (day_data, "day_data")]:
    for col in df.columns:
        if df[col].dtype == object:
            if df[col].dropna().apply(lambda x: str(x).replace('.', '', 1).isdigit()).all():
                df[col] = pd.to_numeric(df[col], errors="ignore")

print("\nUpdated dtypes (hour_data):")
display(hour_data.dtypes.head(20))

print("\nUpdated dtypes (day_data):")
display(day_data.dtypes.head(20))
```

```python
df_day_raw = pd.read_csv('/content/drive/MyDrive/data set/day.csv')

columns_to_drop = ["instant", "dteday", "yr", "casual", "registered"]
day_data_cleaned_df = df_day_raw.drop(columns=columns_to_drop, axis=1)

day_data_cleaned_df.to_csv("day_data_cleaned.csv", index=False)

print("Cleaned day_data saved as day_data_cleaned.csv")

day_data_cleaned_df = pd.read_csv('day_data_cleaned.csv')
day_data_cleaned_df.head()
```

```python
numerical_cols_for_outliers = ['temp', 'atemp', 'hum', 'windspeed', 'cnt']
print("Numerical columns identified for outlier analysis:")
print(numerical_cols_for_outliers)
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols_for_outliers):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=day_data_cleaned_df[col])
    plt.title(f'Box plot of {col}')
    plt.ylabel(col)

plt.tight_layout()
plt.suptitle('Box Plots of Numerical Features (Before Outlier Removal)', y=1.02, fontsize=16)
plt.show()
print("Box plots for numerical features in day_data_cleaned_df before outlier removal displayed.")
```

```python
day_data_no_outliers = day_data_cleaned_df.copy()
outlier_counts = {}

for col in numerical_cols_for_outliers:
    Q1 = day_data_no_outliers[col].quantile(0.25)
    Q3 = day_data_no_outliers[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers before removal
    outliers = day_data_no_outliers[(day_data_no_outliers[col] < lower_bound) | (day_data_no_outliers[col] > upper_bound)]
    outlier_counts[col] = len(outliers)

    # Remove outliers
    day_data_no_outliers = day_data_no_outliers[(day_data_no_outliers[col] >= lower_bound) & (day_data_no_outliers[col] <= upper_bound)]

print("Number of outliers removed per column:")
for col, count in outlier_counts.items():
    print(f"- {col}: {count}")

print(f"\nOriginal DataFrame shape: {day_data_cleaned_df.shape}")
print(f"DataFrame shape after outlier removal: {day_data_no_outliers.shape}")
```

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols_for_outliers):
    plt.subplot(2, 3, i + 1)
    sns.boxplot(y=day_data_no_outliers[col])
    plt.title(f'Box plot of {col} (After Outlier Removal)')
    plt.ylabel(col)

plt.tight_layout()
plt.suptitle('Box Plots of Numerical Features (After Outlier Removal)', y=1.02, fontsize=16)
plt.show()
print("Box plots for numerical features in day_data_no_outliers after outlier removal displayed.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
```

```python
day = pd.read_csv('/content/day_data_cleaned.csv')
```

```python
day.head()
```

```python
X_day = day.drop("cnt", axis=1)
Y_day = day["cnt"]
```

```python
X_train_day, X_test_day, Y_train_day, y_test_day = train_test_split(
    X_day, Y_day, test_size=0.4, random_state=42)
```

```python
print(X_day)
```

```python
print(Y_day)
```

```python
regressor = RandomForestRegressor(n_estimators=100)
```

```python
regressor.fit(X_train_day, Y_train_day)
```

```python
test_day_data_prediction = regressor.predict(X_test_day)
```

```python
print(test_day_data_prediction)
```

```python
error_score = metrics.r2_score(y_test_day, test_day_data_prediction)
print("R squared error : ", error_score)
```

```python
y_test_day = list(y_test_day)
```

```python
plt.plot(y_test_day, color='blue', label = 'Actual Value')
plt.plot(test_day_data_prediction, color='green', label='Predicted Value')
plt.title('Actual  vs Predicted ')
plt.xlabel('Number of values')
plt.ylabel('cnt')
plt.legend()
plt.show()
```
